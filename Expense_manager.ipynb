```python
import datetime

def main():
    # Input Total budget and starting date from the user
    total_budget = float(input("Enter your total budget (in Rs.): "))
    start_date_input = input("Enter the start date (YYYY-MM-DD): ")
    start_date = datetime.datetime.strptime(start_date_input, "%Y-%m-%d").date()
    
    # expenses list
    expenses = []

    print("Enter your expenses one by one (type 'done' when finished):")
    while True:
        expense_name = input("Expense name: ")
        if expense_name.lower() == 'done':
            break
        expense_input = input("Expense amount (in Rs.): ")
        try:
            expense_amount = float(expense_input)
            expenses.append((expense_name, expense_amount))
        except ValueError:
            print("Please enter a valid amount.")

    # Calculating total expenses
    total_expenses = sum(amount for _, amount in expenses)

    # Calculating remaining budget
    remaining_budget = total_budget - total_expenses

    # number of days until the end of the month
    today = datetime.date.today()
    end_of_month = (start_date + datetime.timedelta(days=30)).replace(day=1) + datetime.timedelta(days=-1)
    days_remaining = (end_of_month - start_date).days + 1

    # number of days passed since the start date
    days_passed = (today - start_date).days + 1


    # Calculating average budget per day till the end of the month
    average_budget_per_day = remaining_budget / days_remaining if days_remaining > 0 else 0

    # Calculate expenditure per day
    expenditure_per_day = total_expenses / days_passed if days_passed > 0 else 0

    # Print results
    print("\n--- Expense Summary ---")
    print(f"Total Budget : Rs.{total_budget:.2f}")
    print(f"Total Expenses : Rs.{total_expenses:.2f}")
    print(f"Remaining Budget : Rs.{remaining_budget:.2f}")
    print(f"Average Budget per Day till end of month : Rs.{average_budget_per_day:.2f}")
    print(f"Expenditure per Day : Rs.{expenditure_per_day:.2f}")

    # Print detailed expense report
    print("\n--- Detailed Expenses ---")
    for name, amount in expenses:
        print(f"{name}: Rs.{amount:.2f}")

if __name__ == "__main__":
    main()
```
